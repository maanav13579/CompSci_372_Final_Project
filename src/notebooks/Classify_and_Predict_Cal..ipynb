```python
!pip install -q gradio

import gradio as gr

# Load pipeline
pipeline = FoodCaloriePipeline(
    classifier_path="artifacts/models/classifier_best.pth",
    regressor_path="artifacts/models/regressor_best.pth",
)

def predict(image):
    result = pipeline.predict(image)
    return f"""
    ğŸ½ï¸ **Food:** {result['class_name'].replace('_', ' ').title()}

    ğŸ“Š **Confidence:** {result['confidence']:.1%}

    ğŸ”¥ **Estimated Calories:** {result['calories']:.0f} kcal
    """

# Create interface
demo = gr.Interface(
    fn=predict,
    inputs=gr.Image(type="filepath", label="Upload Food Image"),
    outputs=gr.Markdown(label="Results"),
    title="ğŸ½ï¸ AI Food Recognizer & Calorie Estimator",
    description="Upload a food image to get the predicted food type and calorie estimate.",
)

# Launch (creates a public URL you can share)
demo.launch(share=True)
```
